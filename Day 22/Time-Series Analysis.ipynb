```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

df= pd.read_csv('seattle-weather.csv')
df
```

```python
df.info()
df.shape
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
df.isnull().sum()
```

```python
print("Before Duplicate Values:",df.duplicated().sum())
df.drop_duplicates()
print("After Duplicate Values:",df.duplicated().sum())
```

```python
df = df.sort_values('date')
df.set_index('date', inplace=True)
```

```python
plt.figure(figsize=(10,8))
df['precipitation'].plot()
plt.title('Rainfall Over Time')
plt.xlabel('date')
plt.ylabel('Rainfall')
plt.show()
```

```python
plt.figure(figsize=(10,8))
df['temp_max'].plot()
plt.title("Maximum Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()

plt.figure(figsize=(10,8))
df['temp_min'].plot(kind='hist')
plt.title("Minimum Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
plt.figure(figsize=(10,8))
df['wind'].plot()
plt.title('Wind Over Time')
plt.xlabel('Date')
plt.ylabel('Wind')
plt.show()
```

```python
plt.figure(figsize=(10,8))
df['weather'].value_counts().plot(kind='pie')
plt.title('Weather Type Distribution')
plt.xlabel("Weather Type")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import numpy as np 

df = pd.read_csv("car_price_prediction.csv")
df
```

```python
df[['Brand', 'Year', 'Engine Size', 'Fuel Type', 'Transmission']] = df['Brand,Year,Engine Size,Fuel Type,Transmission'].str.split(
    r',\s*', expand=True)
df
```

```python
df.drop(columns=['Brand,Year,Engine Size,Fuel Type,Transmission'], inplace=True)
df
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ","_")
)
df
```

```python
df.dtypes
```

```python
df['year'] = df['year'].astype('int64')
df['engine_size'] = df['engine_size'].astype('float64')
df.dtypes
```

```python
df.to_csv("Cleaned_Car_Prediction.csv", index=False)
```

```python
import pandas as pd 
import numpy as np 

df = pd.read_csv('messy_dirty_sales_data.csv')
df
```

```python
df = df.sort_values(by='Order_ID')
df
```

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], format='mixed')
df
```

```python
df['Price'] = df['Price'].str.replace('$', '').str.replace(',','').str.replace('₹','').str.replace(r'\D', '', regex=True).astype(float)
df
```

```python
df.info()
df.shape
df.describe()
```

```python
df.describe().T
```

```python
df.replace(99999, np.nan, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['Discount_Code'] = df['Discount_Code'].fillna('NA')
df.isnull().sum()
```

```python
df['Quantity'] = df['Quantity'].fillna(df['Quantity'].median())
df['Price'] = df['Price'].fillna(df['Price'].median())
df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].median())
df.isnull().sum()
```

```python
print("Duplicate Rows:")
print(df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
print("Duplicate Rows:")
print(df.duplicated().sum())
```

```python
df.to_csv('cleaned_sales.csv', index=False)
```

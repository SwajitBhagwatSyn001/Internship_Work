```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Global_Sales_Messy_Data.csv', encoding='utf 8')
df
```

```python
df.dtypes
```

```python
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'], format='mixed')
df
```

```python
from datetime import datetime
df['Server_Timestamp'] = pd.to_datetime(df['Server_Timestamp'],format='mixed', utc=True)
df
```

```python
df['Server_Timestamp'] = df['Server_Timestamp'].dt.strftime('%Y-%m-%d %H:%M:%S') + ' UTC'
df
```

```python
%pip install ftfy
import ftfy
df['Store_Location'] = df['Store_Location'].apply(ftfy.fix_text)
df = df.apply(lambda col: col.apply(ftfy.fix_text) if col.dtype == 'object' else col)
df
```

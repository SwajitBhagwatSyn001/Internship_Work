```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('INC_5000_Companies_2019.csv')
df
```

```python
df.info()
df.shape
df.describe()
```

```python
df.describe().T
```

```python
df.dtypes
```

```python
print("Missing Values")
df.isnull().sum()
```

```python
df['metro'] = df['metro'].fillna('NA')
df['metro'].isnull().sum()
print(df.isnull().sum())
```

```python
df['workers'] = df['workers'].fillna('NA')
df['workers'].isnull().sum()
print(df.isnull().sum())
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df.to_csv("cleaned_companies1.csv",index=False)
```

```python
Industry_count = df['industry'].value_counts().head(10)
print("Industry Count:",Industry_count)
```

```python
# Top 10 Most available industries
plt.figure(figsize = (10,6))
Industry_count.plot(kind='bar', color = 'darkblue')
plt.title("Top 10 Available Industries")
plt.xlabel("Industries")
plt.ylabel("Count of Industries")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# 10 companies with most growth %
avg_growth = df.groupby('industry')['growth_%'].mean().head(10)
plt.figure(figsize=(10,6))
avg_growth.plot(kind='pie', autopct='%1.1f%%', cmap='cool')
plt.title("Top 10 Growth Companies")
plt.xlabel("Industry")
plt.ylabel("Growth %")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Most Companies Established by Year
year_count = df['founded'].value_counts().head(10)
plt.figure(figsize=(10,6))
year_count.plot(kind='bar', color='darkgreen')
plt.title("Companies Most Established by Year")
plt.xlabel("Year")
plt.ylabel("Company Count")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
